```python
# how to handle date of journey column

df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.info()
```

```python
# So here we can see that, the 3 new features that we created are an object type, so we have to convert that features into numerical type
```

```python
#For converting the object types into the numerical type, using 'astype'
df['Date'] = df['Date'].astype('int64')
df['Month'] = df['Month'].astype('int64')
df['Year'] = df['Year'].astype('int64')
```

```python
#Now we can see that the datatpe as been change
df.info()
```

```python
#So now that we have created and converted the date of journey into int(Data, month, year), so there is no need for the data of journey columns. we can drop that column
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
df.info()
```

```python
# Now we will compute arrival time and perform operations on it
#Arrival time have time and dat, but there is no need for date, so we will remove it and split the time
```

```python
# We will now first split the date from the time with the help of lamba and split, spliting with blank space. So as a result we will have only time as output
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(" ")).str[0]
```

```python
# After splitting the data from time, we ahve two int, one before : and one after :, so we will split that using : 
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Min'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df['Arrival_Hour']=df['Arrival_Hour'].astype('int64')
df['Arrival_Min'] = df['Arrival_Min'].astype('int64')
```
